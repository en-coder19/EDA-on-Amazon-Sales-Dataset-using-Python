# Exploratory Data Analysis

```python
da = sns.countplot(x='Size', data=df)
for bars in da.containers:
    da.bar_label(bars)
```

```python
df.groupby(['Size'], as_index = False)['Quantity'].sum().sort_values(by = 'Quantity', ascending = False) 
```

```python
sz_qt = df.groupby(['Size'], as_index = False)['Quantity'].sum().sort_values(by = 'Quantity', ascending = False) 
sns.barplot(x='Size', y='Quantity', data = sz_qt)
```

```python
sns.countplot(data=df, x = 'Courier Status', hue = 'Status')
```

```python
df['Category'] = df['Category']. astype(str)
```

```python
cat = df['Category']
plt.figure(figsize=(10,5))
plt.hist(cat, bins=30)
plt.xticks(rotation = 45)
plt.show()
```

```python
B2B_data = df['B2B'].value_counts()
plt.pie(B2B_data, labels=B2B_data.index, autopct = '%1.1f%%')
plt.title('Representation of type of buyers in terms of percentage')
plt.show()
```

# In the above chart, False refers to Retailers(99.2%), and True refers to B2B buyers(0.8%)

```python
x_data = df['Category']
y_data = df['Size']

plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Availability of sizes of each clothing item')
plt.show()
```

```python
ship_status = df['ship-state'].value_counts().head(10).sort_values(ascending=False)
plt.figure(figsize=(10,5))
sns.countplot(data=df[df['ship-state'].isin(ship_status.index)], x= 'ship-state', order=ship_status.index)
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Top 10 Sates having the highest number of buyers')
plt.xticks(rotation=45)
plt.show()
```
